Goal:
Analyze the relationship between rainfall and crop yield using CSV data.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the Dataset
data = pd.read_csv("/content/agri_data.csv")
```

```python
# Step 2: Show the data
print("Dataset Preview:")
print(data)
```

```python
# Step 3: Data Summary

print("\n Data Summary:")
print(data.describe())
```

```python
# Step 4: Check correlation
correlation = np.corrcoef(data['Rainfall (mm)'], data['Crop Yield (kg/hectare)'])[0, 1]
print(f"\n Correlation between Rainfall and Crop Yield: {correlation:.2f}")
```

```python
# Step 5: Visualization - Line Plot
# Select first 10 rows
data_10 = data.head(10)

plt.figure(figsize=(10, 5))
plt.plot(data['Year'], data['Rainfall (mm)'], marker='o', label='Rainfall (mm)', color='blue')
plt.plot(data['Year'], data['Crop Yield (kg/hectare)'], marker='s', label='Crop Yield (kg/hectare)', color='green')
plt.title("Rainfall vs Crop Yield (Year-wise)")
plt.xlabel("Year")
plt.ylabel("Values")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# Step 6: Visualization - Scatter Plot
plt.figure(figsize=(6, 6))
plt.scatter(data['Rainfall (mm)'], data['Crop Yield (kg/hectare)'], color='purple')
plt.title("Rainfall vs Crop Yield")
plt.xlabel("Rainfall (mm)")
plt.ylabel("Crop Yield (kg/hectare)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Step 7: Average and Maximum Values

# Average
avg_rainfall = np.mean(data['Rainfall (mm)'])
avg_yield = np.mean(data['Crop Yield (kg/hectare)'])

# Maximum
max_rainfall = np.max(data['Rainfall (mm)'])
max_yield = np.max(data['Crop Yield (kg/hectare)'])

print("\n Averages and Maximums:")
print(f"Average Rainfall: {avg_rainfall:.2f} mm")
print(f"Average Crop Yield: {avg_yield:.2f} kg/hectare")
print(f"Maximum Rainfall: {max_rainfall} mm")
print(f"Maximum Crop Yield: {max_yield} kg/hectare")
```

```python
# Step 8 : Line Plot (First 10 Rows)
plt.figure(figsize=(10, 5))
plt.plot(data_10['Year'], data_10['Rainfall (mm)'], marker='*', label='Rainfall (mm)', color='blue')
plt.plot(data_10['Year'], data_10['Crop Yield (kg/hectare)'], marker='s', label='Crop Yield (kg/hectare)', color='g')
plt.title(" Rainfall vs Crop Yield (First 10 Years)")
plt.xlabel("Year")
plt.ylabel("Values")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Step 9: Scatter Plot (First 10 Rows)
plt.figure(figsize=(6, 6))
plt.scatter(data_10['Rainfall (mm)'], data_10['Crop Yield (kg/hectare)'], color='purple', s=80)
plt.title("Rainfall vs Crop Yield (First 10 Years)")
plt.xlabel("Rainfall (mm)")
plt.ylabel("Crop Yield (kg/hectare)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Step 10: Bar Graph for First 10 Rows
plt.figure(figsize=(10, 6))
bar_width = 0.4
x = np.arange(len(data_10['Year']))

plt.bar(x - bar_width/2, data_10['Rainfall (mm)'], width=bar_width, label='Rainfall (mm)', color='skyblue')
plt.bar(x + bar_width/2, data_10['Crop Yield (kg/hectare)'], width=bar_width, label='Crop Yield', color='lightgreen')

plt.xticks(ticks=x, labels=data_10['Year'], rotation=45)
plt.xlabel("Year")
plt.ylabel("Values")
plt.title("First 10 Years: Rainfall vs Crop Yield")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
